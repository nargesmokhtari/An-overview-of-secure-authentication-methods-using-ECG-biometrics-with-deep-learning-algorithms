```python
import os
import shutil

source_folder = "/content/drive/MyDrive/COPY_EKM_dataset_MIT (1)/train"
destination_folder = "/content/drive/MyDrive/trainvalidation/validation/healthy"

selected_names = [
    "5bpf-ekm-MITDB-train-215-",
    "5bpf-ekm-MITDB-train-223-",

]

if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

for filename in os.listdir(source_folder):
    if filename.endswith(".png"):
        for name in selected_names:
            if filename.startswith(name):
                source_path = os.path.join(source_folder, filename)
                destination_path = os.path.join(destination_folder, filename)
                shutil.move(source_path, destination_path)
                print(f"Moved: {filename}")
                break  # No need to check other patterns for this file

```

```python
#  declare the location of our training and validation files
base_dir = '/content/drive/MyDrive/trainvalidation'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')

# cat and dog folders for training
train_cats_dir = os.path.join(train_dir, 'healthy')
train_dogs_dir = os.path.join(train_dir, 'cvd')

# cat and dog folders for validation
validation_cats_dir = os.path.join(validation_dir, 'healthy')
validation_dogs_dir = os.path.join(validation_dir, 'cvd')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Conv2D, MaxPool2D, Flatten, Dense,Dropout
from tensorflow.keras.activations import relu, sigmoid
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import binary_accuracy
import tensorflow as tf
# create the model
model = Sequential()
# Add the hidden layers
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(21, 33, 3)))
model.add(MaxPool2D(pool_size=(2, 2), strides=None, padding='valid'))
model.add(Dropout(0.7))
# Add the flatten layer
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python

from tensorflow.keras.preprocessing.image import ImageDataGenerator

# we rescale all our images with the rescale parameter
train_datagen = ImageDataGenerator(rescale = 1.0/255)
test_datagen  = ImageDataGenerator(rescale = 1.0/255)

# we use flow_from_directory to create a generator for training
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    batch_size=32,
                                                    class_mode='binary',
                                                    target_size=(21, 33))

# we use flow_from_directory to create a generator for validation
validation_generator =  test_datagen.flow_from_directory(validation_dir,
                                                         batch_size=32,
                                                         class_mode='binary',
                                                         target_size=(21, 33))
```

```python
history = model.fit(
            train_generator, # pass in the training generator
            steps_per_epoch=100,
            epochs=20,
            validation_data=validation_generator, # pass in the validation generator
            validation_steps=50,
            verbose=2
            )
```

```python
import os
from keras.preprocessing import image
from PIL import Image

# Directory containing the PNG images
image_directory = '/content/drive/MyDrive/COPY_NSRDB_dataset_MIT/test'

# List all files in the directory
image_files = os.listdir(image_directory)

# Initialize counters for dogs and cats
healthy_count = 0
cvd_count = 0

for fn in image_files:
    if fn.endswith('.png'):  # Process only PNG images
        # Construct the full path to the image
        path = os.path.join(image_directory, fn)

        # Load and preprocess the image
        img = Image.open(path)
        img = img.resize((33, 21))
        img = img.convert('RGB')


        x = np.array(img)/255.0
        x = np.expand_dims(x, axis=0)

        # Make prediction
        classes = model.predict(x, batch_size=1)

        if classes[0] > 0.5:
            print(fn + " healthy")
            healthy_count += 1
        else:
            print(fn + " cvd!")
            cvd_count += 1

# Print the total count for each class
print("Total healthy count:", healthy_count)
print("Total cvd count:", cvd_count)
```
